```python
def labels_classes(labels):
    # find the unique class labels
    return np.unique(labels)

def labels_to_one_hots(labels,classes):
    num_labels = labels.shape[0]
    num_classes = classes.shape[0]

    # create index of label into the hots..
    #   ie covert the label to a number indexing into classes
    label_index = np.zeros(labels.shape[0], dtype=int)
    idx = 0
    for c in classes:
        label_index[labels == c] = idx
        idx += 1

    # compute the offsets into the one_hots (if it was flat)
    hot_indexes = np.arange(num_labels) * num_classes + label_index

    # and then write each item that is a hot as a hot
    one_hots = np.zeros((num_labels, num_classes))
    one_hots.flat[hot_indexes] = 1

    return one_hots

def xy_range(xy):
    xy = xy.astype(float)
    
    xy_max = np.amax(xy,axis=0)
    xy_min = np.amin(xy,axis=0)

    return xy_min, xy_max

def xy_norm(xy, xy_min, xy_max):
    xy = xy.astype(float)
    
    return (xy - xy_min)/(xy_max - xy_min)
```

```python
xy_min, xy_max = xy_range(clean[['X','Y']].values)
xys = xy_norm(clean[['X','Y']].values, xy_min, xy_max)
classes = labels_classes(clean["Category"].values)
cats  = labels_to_one_hots(clean["Category"].values, classes)

print xy_min, xy_max
print xys
print classes
print cats

print np.sum(cats, axis=0)
```

```python
def create_tfidf(xys, cats, classes):
    # de-one hot
    size = 100
    class_count = classes.shape[0]
    class_idents = np.argmax(cats, axis=1).reshape((cats.shape[0],1))
    
    all = np.ones((size+1,size+1))
    images = np.ones((class_count,size+1,size+1))

    for c in range(class_count-1):
        idx = (class_idents == c).reshape(class_idents.shape[0])
        selection = xys[idx]

        for i in range(selection.shape[0]):
            x = int(size*selection[i,0])
            y = int(size*selection[i,1])

            images[c,x,y] += 1  # all images
            all[x,y] += 1
    
    tfidf = np.log(all) - np.log(all/images)

    if not os.path.exists("tfidf"):
        os.makedirs("tfidf")
        
    for c in range(class_count-1):
        label = classes[c]
        label = re.sub(' ', '_', label)
        label = re.sub('\/', '_', label)
        label = "tfidf/" + label + "_tfidfl.png"

        fig = plt.figure()
        ax = fig.add_subplot(1, 1, 1)
        ax.set_title(classes[c] + " @ tfidf")
        y_dim = tfidf.shape[2]
        x_dim = tfidf.shape[1]
        im = images[c,:,:].reshape(y_dim, x_dim)
        ax.imshow(im, cmap='gist_heat', interpolation='nearest')
        ax.axis('off')
        plt.savefig(label)
        plt.show()
        plt.close(fig)

    return tfidf

def build_softpdf(tfidf):
    exp_tfidf = np.exp(tfidf)
    softmax_sum = np.sum(exp_tfidf, axis=0)
    softmax = exp_tfidf / softmax_sum

    most_likely = np.argmax(softmax,axis=0)

    label = "tfidf/most_likely_tfidfl.png"
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.set_title("most likely tfidf")
    y_dim = most_likely.shape[1]
    x_dim = most_likely.shape[0]
    im = most_likely.reshape(y_dim, x_dim)
    ax.imshow(im, cmap='Set1', interpolation='nearest')
    ax.axis('off')
    plt.savefig(label)
    plt.show()
    plt.close(fig)

    return softmax
```

```python
tfidf = create_tfidf(xys, cats, classes)
softmax = build_softpdf(tfidf)
```

```python
def logloss(act, pred):
    epsilon = 1e-15
    pred = np.maximum(  epsilon, pred)
    pred = np.minimum(1-epsilon, pred)
    ll = -1.0 * np.mean(act*np.log(pred) + (1 - act)*np.log(1 - pred), axis=0)
    return np.sum(ll)

def run_model(softmax, test_in, test_out):
    results = np.zeros((test_in.shape[0], tfidf.shape[0]))
    for i in range(test_in.shape[0]):
        x = np.maximum(0, np.minimum(100, int(100*test_in[i,0])))
        y = np.maximum(0, np.minimum(100, int(100*test_in[i,1])))
        res = softmax[:,x,y]

        results[i,:] = res

    if (not test_out is None):
        correct_prediction = np.argmax(results,axis=1) == np.argmax(test_out,axis=1)
        accuracy = float(np.sum(correct_prediction)) / float(correct_prediction.shape[0])

        lloss = logloss(test_out,results)

        print " correct:", np.sum(correct_prediction), "/", correct_prediction.shape[0]
        print " tfidf accuracy:", accuracy, "log loss", lloss

    return results
```
